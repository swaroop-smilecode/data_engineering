##### <ins>pandas can't automatically interpret the data type of column, just by reading the data</ins>
That's the reason, see below. data type of column named 'Date/Time' is reflected as an object.<br>

```python
weather_df['Date/Time'].head()
```

So, we need to explicilty teach pandas about the data type, before doing any further operations on that column.<br>
Let's convert the `Date/Time` column datatype from object to `timestamp` so that we can access the month directly using the attribute `dt.month`

```python
weather_df['Date/Time'] = pd.to_datetime(weather_df['Date/Time'])
weather_df['Date/Time'].head()
```
